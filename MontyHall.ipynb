### Question 1: Pen-and-paper exercise

This is what is refered to as the Monty Hall Problem. When you initially choose a door, there is a 1/3 chance that the car is behind that door and a 2/3 chance that the car is behind one of the other two doors.

When the host, who knows what’s behind each door, opens another door revealing a goat, the probabilities don’t change. There is still a 1/3 chance that the car is behind your initial choice, and a 2/3 chance that the car is behind one of the other two doors. But since one of those doors has been opened to reveal a goat, the entire 2/3 probability now rests with the remaining unopened door.

So, if you switch doors after the host reveals a goat, your probability of winning the car increases from 1/3 to 2/3. If you stick with your initial choice, your probability remains at 1/3. Therefore, it is to your advantage to switch your choice of doors.

### Question 2: Simulation in Python

```python
import random as rand

samples = 200
cars_won = 0
in_door_1 = 0

for x in range(0, samples):
    doors = ["goat", "goat", "goat"]
    
    doors[rand.randint(0,2)] = "car"
    
    print("Game:", x)
    print("You picked door 1:", doors[0])
    if doors[0] == "car":
        in_door_1 += 1
    
    if doors[1] == "goat":
        print("Host opens door 2 and shows a goat!")
        print("You switch to door 3 and get a", doors[2] + "!")
        picked = doors[2]
    else:
        print("Host opens door 3 and shows a goat!")
        print("You switch to door 2 and get a", doors[1] + "!")
        picked = doors[1]

    if picked == "car":
        cars_won += 1
        

print("Cars won from staying:", format(in_door_1/samples, ".0%"))
print("Cars won from switching:", format(cars_won/samples, ".0%"))
```
